```python
# ===============================
# FRAUD DETECTION 
# ===============================

import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import (
    classification_report,
    confusion_matrix,
    roc_auc_score,
    precision_recall_curve
)
```

```python
# -------------------------------
# 1) LOAD DATA
# -------------------------------
df = pd.read_csv("creditcard.csv")
print("Dataset shape:", df.shape)
print("Fraud rate:", df["Class"].mean())
```

```python
import matplotlib.pyplot as plt

df['Class'].value_counts().plot(kind='bar')
plt.title("Class Distribution (0 = Legit, 1 = Fraud)")
plt.show()

print(df['Class'].value_counts(normalize=True))
```

```python
plt.figure(figsize=(8,4))
sns.histplot(df[df['Class']==0]['Amount'], bins=50, label='Legit', color='blue', stat='density')
sns.histplot(df[df['Class']==1]['Amount'], bins=50, label='Fraud', color='red', stat='density')
plt.legend()
plt.yscale('log')
plt.title("Amount Distribution")
plt.show()
```

```python
features = ['V1', 'V2', 'V3']

for f in features:
    plt.figure(figsize=(6,3))
    sns.kdeplot(df[df['Class']==0][f], label='Legit')
    sns.kdeplot(df[df['Class']==1][f], label='Fraud')
    plt.title(f"Distribution of {f}")
    plt.legend()
    plt.show()
```

```python
# ================================
# FEATURE ENGINEERING (AMOUNT-BASED)
# ================================

# 1 Log transform of Amount (reduces skew)
df["log_amount"] = np.log1p(df["Amount"])

# 2 High amount flag (top 1%)
high_amt_threshold = df["Amount"].quantile(0.99)
df["is_high_amount"] = (df["Amount"] > high_amt_threshold).astype(int)


# 3 Interaction features (Amount  PCA components)
df["amount_V1"] = df["Amount"] * df["V1"]
df["amount_V2"] = df["Amount"] * df["V2"]

print("New features added:")
print([
    "log_amount",
    "is_high_amount",
    "amount_V1",
    "amount_V2"
])
```

```python
# -------------------------------
# 2) FEATURES / TARGET
# -------------------------------
X = df.drop(columns=["Class"])
y = df["Class"]
```

```python
# -------------------------------
# 3) TRAINTEST SPLIT (STRATIFIED)
# -------------------------------
X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.2,
    stratify=y,
    random_state=42
)
```

```python
# -------------------------------
# 4) SCALING (IMPORTANT FOR LR)
# -------------------------------
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled  = scaler.transform(X_test)

# -------------------------------
# 5) BASELINE MODEL (IMBALANCE-AWARE)
# -------------------------------
lr = LogisticRegression(
    max_iter=1000,
    class_weight={0: 1.0, 1: 10.0},  # penalize missing fraud
    solver="liblinear"
)

lr.fit(X_train_scaled, y_train)
```

```python
# -------------------------------
# 6) PROBABILITIES (NOT .predict)
# -------------------------------
y_prob = lr.predict_proba(X_test_scaled)[:, 1]

print("ROCAUC:", roc_auc_score(y_test, y_prob))

# -------------------------------
# 7) THRESHOLD TUNING
# -------------------------------
print("\nThreshold sweep:\n")

for t in [0.01, 0.02, 0.05, 0.10]:
    y_pred = (y_prob >= t).astype(int)
    print(f"Threshold = {t}")
    print(confusion_matrix(y_test, y_pred))
    print(classification_report(y_test, y_pred, digits=4))
    print("-" * 50)
```

```python
# -------------------------------
# 8) FINAL CHOICE (BUSINESS-DRIVEN)
# -------------------------------
FINAL_THRESHOLD = 0.23
final_pred = (y_prob >= FINAL_THRESHOLD).astype(int)

print("\nFINAL MODEL @ Threshold =", FINAL_THRESHOLD)
print(confusion_matrix(y_test, final_pred))
print(classification_report(y_test, final_pred, digits=4))

# -------------------------------
# 9) PRECISIONRECALL CURVE (OPTIONAL PLOT)
# -------------------------------
precision, recall, thresholds = precision_recall_curve(y_test, y_prob)

print("\nSample PR points:")
for p, r in zip(precision[:5], recall[:5]):
    print(f"Precision={p:.3f}, Recall={r:.3f}")
```

```python
# ================================
# BUSINESS COST ANALYSIS
# ================================

# Confusion matrix values from FINAL MODEL
TN, FP, FN, TP = confusion_matrix(y_test, final_pred).ravel()

# Cost assumptions (in INR)
COST_FN = 10000   # fraud missed
COST_FP = 500     # customer inconvenience

total_fraud_loss = FN * COST_FN
total_fp_cost = FP * COST_FP
total_cost = total_fraud_loss + total_fp_cost

print("BUSINESS COST ANALYSIS")
print("------------------------")
print(f"False Negatives (Frauds Missed): {FN}")
print(f"False Positives (False Alerts): {FP}")
print()
print(f"Fraud Loss        : {total_fraud_loss:,}")
print(f"Customer Friction : {total_fp_cost:,}")
print(f"TOTAL COST        : {total_cost:,}")
```

```python
# ================================
# SAVE PRODUCTION ARTIFACTS
# ================================

import joblib

joblib.dump(lr, "fraud_model.pkl")
joblib.dump(scaler, "scaler.pkl")
joblib.dump(X.columns.tolist(), "feature_columns.pkl")
joblib.dump(FINAL_THRESHOLD, "threshold.pkl")

print("Artifacts saved:")
print("- fraud_model.pkl")
print("- scaler.pkl")
print("- feature_columns.pkl")
print("- threshold.pkl")


```
